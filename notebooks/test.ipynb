```python
import sys
from random import randint
prog = sys.argv[1]

transition = dict()

with open('prog2') as f:
    while l := f.readline():
        if l == '\n':
            continue
        l = l.strip("\n").replace(" ", "").split('#')[0].split('->')
        src = l[0].split(',')
        dst = l[1].split(',')
        if src[0] not in transition:
            transition[src[0]] = dict()
        aux = transition[src[0]]
        if len(src) == 1:
            aux[None] = dst
        else:
            aux[src[1]] = dst           

for it in range(1):
    initial = '1' * randint(1, 10) + '0' * randint(1, 10)
    head = 0
    initial = "11111000"
    tape = list(initial)
    state = "S0"

    for i in range(head):
        print(' ', end="")
    print('v')
    print(f'{"".join(tape)} {state}')


    while state in transition:
        if head < 0:
            tape.insert(0, '_')
        elif head >= len(tape):
            tape.append('_')
        t = transition[state]
        if None in t:
            if t[None][0] == 'R':
                head += 1
                state = t[None][1]
            elif t[None][0] == 'L':
                head -= 1
                state = t[None][1]
        else:
            if tape[head] in t:
                state = t[tape[head]][1]
                tape[head] = t[tape[head]][0]
            else:
                break
        for i in range(head):
            print(' ', end="")
        print('v')
        print(f'{"".join(tape)} {state}')

    final = "".join(tape).strip('_')
    print(final)
    ok = (final == initial[::-1])
    if not ok:
        print("Not OK")
        break
```
