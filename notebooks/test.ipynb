```python
import minari

dataset = minari.load_dataset("pointmaze/umaze-v1")

# Load all episodes from the dataset
episodes = list(dataset.iterate_episodes())

print(f"Loaded dataset with {len(episodes)} episodes.")
```

```python
import matplotlib.pyplot as plt

for i, ep in enumerate(episodes[:5]):  # First 5 episodes
    qpos = ep.infos["qpos"]  # Agent positions
    x = [p[0] for p in qpos]
    y = [p[1] for p in qpos]

    plt.plot(x, y, label=f"Episode {i+1}")

plt.title("Trajectories in U-Maze")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis("equal")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import sys
from random import randint
prog = sys.argv[1]

transition = dict()

with open('prog2') as f:
    while l := f.readline():
        if l == '\n':
            continue
        l = l.strip("\n").replace(" ", "").split('#')[0].split('->')
        src = l[0].split(',')
        dst = l[1].split(',')
        if src[0] not in transition:
            transition[src[0]] = dict()
        aux = transition[src[0]]
        if len(src) == 1:
            aux[None] = dst
        else:
            aux[src[1]] = dst           

for it in range(1):
    initial = '1' * randint(1, 10) + '0' * randint(1, 10)
    head = 0
    initial = "11111000"
    tape = list(initial)
    state = "S0"

    for i in range(head):
        print(' ', end="")
    print('v')
    print(f'{"".join(tape)} {state}')


    while state in transition:
        if head < 0:
            tape.insert(0, '_')
        elif head >= len(tape):
            tape.append('_')
        t = transition[state]
        if None in t:
            if t[None][0] == 'R':
                head += 1
                state = t[None][1]
            elif t[None][0] == 'L':
                head -= 1
                state = t[None][1]
        else:
            if tape[head] in t:
                state = t[tape[head]][1]
                tape[head] = t[tape[head]][0]
            else:
                break
        for i in range(head):
            print(' ', end="")
        print('v')
        print(f'{"".join(tape)} {state}')

    final = "".join(tape).strip('_')
    print(final)
    ok = (final == initial[::-1])
    if not ok:
        print("Not OK")
        break
```
