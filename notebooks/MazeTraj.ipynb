```python
import minari
import torch
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
```

```python

# Need to run 'minari download D4RL/pointmaze/large-dense-v2'
dataset = minari.load_dataset('D4RL/pointmaze/large-dense-v2')
env  = dataset.recover_environment()
```

```python
for episode in dataset:
    print(f"EPISODE ID {episode.id}")
```

```python

# Extract all episodes
obs_list, action_list = [], []
for episode in dataset:
    obs_list.append(episode.observations)   # States
    action_list.append(episode.actions)     # Actions

 
```

```python
   
print(f"Number of episodes: {len(action_list)}")
print(action_list[0])
```

```python
# Print the first observation
print("First observation:", obs_list[0])
print("Observation keys:", obs_list[0].keys() if isinstance(obs_list[0], dict) else "Not a dictionary")

# Print the first action
print("First action:", action_list[0])
print("Action keys:", action_list[0].keys() if isinstance(action_list[0], dict) else "Not a dictionary")
```

```python
#x, y and, observation
print(obs_list[0]['observation'])
```

```python
# Process observations (trajectory)
obs_list_fixed = []
for obs in obs_list:
    if isinstance(obs, dict):
        obs = obs['observation']  # Adjust key if necessary
    obs_list_fixed.append(np.array(obs, dtype=np.float32))

# Convert list of observations into a single NumPy array
obs_array = np.vstack(obs_list_fixed)

# Convert to tensor
obs_tensor = torch.tensor(obs_array, dtype=torch.float32)

# Create TensorDataset (only observations)
maze_dataset = TensorDataset(obs_tensor)

# Create DataLoader
dataloader = DataLoader(maze_dataset, batch_size=128, shuffle=True)

# Print a sample batch
for obs_batch in dataloader:
    print("Observation Batch Shape:", obs_batch[0].shape)  # Indexing because TensorDataset returns a tuple
    print(obs_batch[0])
    break
```
