```python
import minari
import torch
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
```

```python

# Need to run 'minari download D4RL/pointmaze/large-dense-v2'
dataset = minari.load_dataset('D4RL/pointmaze/large-dense-v2')
env  = dataset.recover_environment()
```

```python
for episode in dataset:
    print(f"EPISODE ID {episode.id}")
```

```python

# Extract all episodes
obs_list, action_list = [], []
for episode in dataset:
    obs_list.append(episode.observations)   # States
    action_list.append(episode.actions)     # Actions

 
```

```python
   
print(f"Number of episodes: {len(action_list)}")
print(action_list[0])
```

```python
# Convert observations to a uniform shape
obs_list_fixed = []
for obs in obs_list:
    obs = np.array(obs, dtype=np.float32)  # Convert each observation to a NumPy array
    obs_list_fixed.append(obs)

# Stack all observations into a single NumPy array
obs_array = np.vstack(obs_list_fixed)


# Convert to tensors
obs_tensor = torch.tensor(obs_array, dtype=torch.float32)
action_tensor = torch.tensor(action_array, dtype=torch.float32)

# Create DataLoader
maze_dataset = TensorDataset(obs_tensor, action_tensor)
dataloader = DataLoader(maze_dataset, batch_size=128, shuffle=True)

# Print a sample batch
for obs_batch, action_batch in dataloader:
    print("Observation Batch Shape:", obs_batch.shape)
    print("Action Batch Shape:", action_batch.shape)
    break
```
